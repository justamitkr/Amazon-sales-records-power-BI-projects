### Step 1: Removing duplicate rows ( there could be duplicate rows excluding Row_ID column ). 

```python
#duplicate data removal
ss_data = store_data.drop_duplicates()
```

```python
#Reading the new dataframe
ss_data.head()
```

```python
ss_data.info()
```

#### There are no duplicate values

### Step 2: Removing rows for which few values are missing.

#### Only postal code has 11 missing values. Lets remove it as required.

```python
#dropping columns
ss_data.drop('Postal_Code', axis=1, inplace=True)
```

```python
ss_data.columns
```

### Step 3: Remove irrelevant values from each column if any. Validation of all values for a column( order date and ship date value must be in correct date format ). For each entry in dataset ship date >= order date 

```python
# Convert dates to datetime format
ss_data['Ship_Date'] = pd.to_datetime(ss_data['Ship_Date'])
ss_data['Order_Date'] = pd.to_datetime(ss_data['Order_Date'])
```

```python
# Create a new column 'ship_after_order' that is True if the ship date is >= order date, otherwise False
ss_data['ship_after_order'] = ss_data['Ship_Date'] >= ss_data['Order_Date']
```

```python
#checking unique values
ss_data['ship_after_order'].unique()
```

#### The above result shows that there are rows where ship date is < order date which is absurd. 

```python
# Get the count of unique values in a column
unique_count = ss_data['ship_after_order'].nunique()
```

```python
# Get the count of how many times a value appears in a column
value_count = ss_data['ship_after_order'].value_counts()
```

```python
# Print the value counts
print(value_count)
```

#### It can be clearly observed that there are 1684 instances where the data for ship date and order date is absurd.

#### These rows need to be removed because it contains absurd/irrelevant/wrong values.

```python
# Drop rows where a specific column value equals a certain value
df = ss_data.drop(ss_data[ss_data['ship_after_order'] == False].index)
```

```python
# Print the updated DataFrame
df.info()
```

```python
#verifying data
df['ship_after_order'].unique()
```

### Step 4: Export the cleaned dataset as a .csv file: prefer UTF-8 encoding. 

```python
# Export data to a CSV file with UTF-8 encoding
df.to_csv('cleaned_superstore_data.csv', encoding='utf-8', index=False)
```
